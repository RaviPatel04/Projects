```python
import pandas as pd
import os

# Specify the folder containing the CSV files
folder_path = 'D:\web_series_recommendation'

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Initialize an empty DataFrame
combined_df = pd.DataFrame()

# Loop through each file and append it to the DataFrame
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('combined_output.csv', index=False)
```

```python
# Function to extract and clean Stars
def extract_and_clean_stars(cast_info):
    # Split the content by "Stars:"
    parts = cast_info.split('Stars:,')
    
    if len(parts) > 1:
        # Get the part after "Stars:"
        stars_part = parts[1].split('|')[0].strip()
        
        # Remove extra colons and clean up the string
        stars_part = stars_part.replace(',,', ',').strip()
        # Optionally, remove any leading/trailing commas
        stars_part = ', '.join([star.strip() for star in stars_part.split(',') if star.strip()])
        
        return stars_part
    return ''
```

```python
# Apply the function to the 'cast' column
df['Stars'] = df['Cast'].apply(extract_and_clean_stars)

df
```

```python
def convert_to_list(stars_string):
    # Split the string by commas and clean up each entry
    stars_list = [star.strip() for star in stars_string.split(',') if star.strip()]
    return stars_list
```

```python
df['Stars'] = df['Stars'].apply(convert_to_list)
```

```python
df['Genre'] = df['Genre'].apply(convert_to_list)
```

```python
df['Synopsis'] = df['Synopsis'].apply(lambda x:x.split())
```

```python
df.head(1)
```

```python
df['tags'] = df['Genre'] + df['Stars'] + df['Synopsis']
```

```python
df.drop(columns=['Genre','Stars','Synopsis','Cast'], inplace=True)
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df.head(2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# TF-IDF Vectorization
tfidf = TfidfVectorizer(max_features=2000, stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['tags'])

# Nearest Neighbors
nn = NearestNeighbors(n_neighbors=6, metric='cosine', algorithm='brute')  # or use 'ball_tree'
nn.fit(tfidf_matrix)
```

```python
# Suggest Function using Nearest Neighbors
def suggest(movie_name):
    index = df[df['Title'] == movie_name].index[0]
    distances, indices = nn.kneighbors(tfidf_matrix[index], n_neighbors=6)
    
    for i in indices[0][1:]:
        print(df.iloc[i].Title)
```

```python
suggest("The World at War")
```
