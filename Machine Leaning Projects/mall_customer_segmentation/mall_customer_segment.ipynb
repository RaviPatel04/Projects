```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("mall_customers.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,[3,4]].values
```

```python
X
```

```python
# WCSS -> Within Clusters Sum of Squares

wcss = []
for i in range(1,11):
    clusters = KMeans(n_clusters=i, random_state=20,init='k-means++')
    clusters.fit(X)
    wcss.append(clusters.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
cluster = KMeans(n_clusters=5, random_state=20, init='k-means++')
y = cluster.fit_predict(X)
y
```

```python
sns.set_theme(style='darkgrid')
plt.scatter(X[y==0,0],X[y==0,1],s=20,c="blue",label = "cluster-1")
plt.scatter(X[y==1,0],X[y==1,1],s=20,c="green",label = "cluster-1")
plt.scatter(X[y==2,0],X[y==2,1],s=20,c="red",label = "cluster-1")
plt.scatter(X[y==3,0],X[y==3,1],s=20,c="cyan",label = "cluster-1")
plt.scatter(X[y==4,0],X[y==4,1],s=20,c="magenta",label = "cluster-1")

plt.scatter(cluster.cluster_centers_[:,0], cluster.cluster_centers_[:,1], s=50, c='black', label='Centroids',marker="*")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
