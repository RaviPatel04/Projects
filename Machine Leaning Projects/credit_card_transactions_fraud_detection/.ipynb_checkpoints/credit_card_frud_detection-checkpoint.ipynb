```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df = pd.read_csv("fraudtrain.csv")
```

```python
not_fraud = df[df.is_fraud == 0]
fraud = df[df.is_fraud == 1]
```

```python
print(not_fraud.shape)
print(fraud.shape)
```

```python
not_fraud_sample = not_fraud.sample(n=7506,random_state = 4)
```

```python
new_df = pd.concat([not_fraud_sample,fraud])
```

```python
new_df
```

```python
useless_cols = [
    'Unnamed: 0',
    'merchant',
    'cc_num',
    'first',
    'zip',
    'last',
    'trans_num',
    'unix_time',
    'street',
    'merch_lat',
    'gender',
    'merch_long',
    'job',
    'trans_date_trans_time'
]

# Dropping the columns
new_df.drop(columns=useless_cols, inplace=True)
```

```python
new_df
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
encoder=LabelEncoder()
new_df['category']=encoder.fit_transform(new_df['category'])
new_df['city']=encoder.fit_transform(new_df['city'])
new_df['state']=encoder.fit_transform(new_df['state'])
new_df['dob']=encoder.fit_transform(new_df['dob'])
```

```python
new_df
```

```python
plt.figure(figsize = (12, 12))
sns.heatmap(new_df.corr(), annot = True,cmap="Blues")
plt.show()
```

```python
X = new_df.drop(columns="is_fraud",axis =1)
X
```

```python
y = new_df["is_fraud"]
y
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pred_train = model.predict(X_train)
acc_train = accuracy_score(pred_train,y_train)
acc_train
```

```python
pred_test = model.predict(X_test)
acc_test = accuracy_score(pred_test,y_test)
acc_test
```

```python
import joblib
joblib.dump(scaler, "credit_card_fraud_scaler.joblib")
joblib.dump(model, "credit_card_fraud_model.joblib")
```
