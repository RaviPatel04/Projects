### The Impact of Parental Education on Student Performance

```python
pe_effect = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
pe_effect
```

```python
pe_effect_sorted = pe_effect.sort_values(by="MathScore", ascending=False)
pe_effect_sorted
```

```python
sns.heatmap(pe_effect_sorted, annot = True)
plt.show()
```

Parent education positively correlates with student performance, with higher-educated parents generally associated with better Math, Reading, and Writing scores.
